Get column comments

```python
QUERY = """
SELECT c.table_schema,c.table_name,c.column_name,pgd.description
FROM pg_catalog.pg_statio_all_tables as st
  inner join pg_catalog.pg_description pgd on (pgd.objoid=st.relid)
  right join information_schema.columns c on (pgd.objsubid=c.ordinal_position
    and  c.table_schema=st.schemaname and c.table_name=st.relname)
where c.table_schema not in ('information_schema', 'pg_catalog', 'public')
"""


with engine.begin() as conn:
    df_columns = pd.read_sql(QUERY, conn)
```

Write to excel

```python
writer = pd.ExcelWriter('data_dictionary.xlsx')
df_tables.to_excel(writer,'tables')
df_columns.to_excel(writer,'columns')

writer.save()
    
    
```
